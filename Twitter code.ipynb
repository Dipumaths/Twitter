```python
# Twitter Sentiment Analysis - Complete Notebook
# Paste into a Jupyter cell (or split into multiple cells)

# ========================
# 0. Notes / Requirements
# ========================
# - Make sure you have installed required packages:
#   pip install numpy pandas matplotlib seaborn nltk scikit-learn wordcloud tensorflow keras xgboost tqdm
# - Replace 'Tweets.csv' path below with the correct local path.
# - This notebook trains several models; training DL models may take time depending on CPU/GPU.

# ========================
# 1. Imports & Setup
# ========================
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Display settings
pd.set_option('display.max_colwidth', 200)
%matplotlib inline

# NLP & preprocessing
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Classic ML
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier

# Deep learning
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Warnings
import warnings
warnings.filterwarnings('ignore')

# ========================
# 2. Load Dataset
# ========================
# Change this path to where your CSV is located
DATA_PATH = r"C:\Users\dipum\Downloads\Tweets csv.csv"

if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATA_PATH}. Please place Tweets.csv in the working directory or update DATA_PATH variable.")

df = pd.read_csv(DATA_PATH)

# Preview and basic cleanup based on your PDF
if 'selected_text' in df.columns:
    df = df.drop(columns=['selected_text'], errors='ignore')
if 'textID' in df.columns:
    df = df.drop(columns=['textID'], errors='ignore')

# Drop null rows, reset index
df = df.dropna(subset=['text', 'sentiment']).reset_index(drop=True)
print("Dataset shape:", df.shape)
df.head()

# ========================
# 3. Quick EDA
# ========================
print("\nSentiment distribution:")
print(df['sentiment'].value_counts())

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sentiment')
plt.title("Sentiment Counts")
plt.show()

# Basic text stats
df['word_count'] = df['text'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['text'].apply(lambda x: len(str(x)))
df['avg_word_len'] = df['text'].apply(lambda x: np.mean([len(w) for w in str(x).split()]) if len(str(x).split())>0 else 0)
df[['word_count','char_count','avg_word_len']].describe()

# ========================
# 4. WordClouds (optional)
# ========================
from wordcloud import WordCloud, STOPWORDS
def plot_wordcloud(text_series, title="WordCloud"):
    words = ' '.join(text_series.astype(str))
    cleaned = " ".join([w for w in words.split() if 'http' not in w and not w.startswith('@') and w!='RT'])
    wc = WordCloud(stopwords=STOPWORDS, background_color='black', width=1600, height=600).generate(cleaned)
    plt.figure(figsize=(14,5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

plot_wordcloud(df[df['sentiment']=='negative']['text'], 'Negative WordCloud')
plot_wordcloud(df[df['sentiment']=='positive']['text'], 'Positive WordCloud')
plot_wordcloud(df[df['sentiment']=='neutral']['text'], 'Neutral WordCloud')

# ========================
# 5. Preprocessing Functions
# ========================
stop = set(stopwords.words('english'))
ps = PorterStemmer()

def clean_text(text, remove_stopwords=True, do_stem=True):
    text = str(text)
    # Lowercase
    text = text.lower()
    # Remove urls
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove @mentions and RT
    text = re.sub(r'@\w+', '', text)
    text = text.replace('rt', '')
    # Remove punctuation and non-alphanumeric (keep spaces)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    # Collapse multiple spaces
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    if remove_stopwords:
        tokens = [t for t in tokens if t not in stop]
    if do_stem:
        tokens = [ps.stem(t) for t in tokens]
    return " ".join(tokens)

# Apply cleaning (may take a short while)
tqdm.pandas(desc="Cleaning texts")
df['clean_text'] = df['text'].progress_apply(lambda x: clean_text(x))

# Quick peek
df[['text','clean_text']].head(8)

# ========================
# 6. Prepare labels
# ========================
le = LabelEncoder()
df['label'] = le.fit_transform(df['sentiment'])  # e.g. negative=0, neutral=1, positive=2
num_classes = len(le.classes_)
print("Classes:", le.classes_)

# ========================
# 7. Train/Test split for classical models (TF-IDF)
# ========================
X_tfidf = df['clean_text']
y = df['sentiment']  # keep string labels for sklearn metrics

tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,2))
X_tfidf_mat = tfidf.fit_transform(X_tfidf)

X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(
    X_tfidf_mat, y, test_size=0.2, random_state=42, stratify=df['label']
)
print("TF-IDF matrix shape:", X_tfidf_mat.shape)
print("Train/Test sizes:", X_train_tfidf.shape, X_test_tfidf.shape)

# ========================
# 8. Evaluation Helper (multiclass)
# ========================
results = []

def evaluate_and_store(name, y_true, y_pred, y_score=None):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    # AUC: for multiclass, require one-hot true & predicted probabilities
    auc = None
    if y_score is not None:
        try:
            y_true_ohe = pd.get_dummies(y_true, columns=le.classes_)
            # Ensure order of classes same as le.classes_
            auc = roc_auc_score(y_true_ohe.values, y_score, multi_class='ovr')
        except Exception:
            auc = None
    results.append({
        'Model': name,
        'Accuracy': round(acc*100,3),
        'Precision': round(prec*100,3),
        'Recall': round(rec*100,3),
        'F1': round(f1*100,3),
        'AUC': round(auc*100,3) if auc is not None else None
    })
    print(f"\n{name} results:")
    print(f"Accuracy: {acc:.4f}  F1 (weighted): {f1:.4f}")
    print("Classification report:\n", classification_report(y_true, y_pred, zero_division=0))

# ========================
# 9. Train classical ML models
# ========================
sk_models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Naive Bayes (BernoulliNB)": BernoulliNB(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM (RBF, prob=True)": SVC(probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=150, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

for name, model in sk_models.items():
    print("\nTraining:", name)
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    y_score = None
    if hasattr(model, "predict_proba"):
        try:
            y_score = model.predict_proba(X_test_tfidf)
        except Exception:
            y_score = None
    evaluate_and_store(name, y_test, y_pred, y_score)

# Show results table for classic models
res_df = pd.DataFrame(results)
display(res_df.sort_values('F1', ascending=False))

# ========================
# 10. Prepare data for Deep Learning (Tokenizer + Embedding)
# ========================
# We'll use the clean_text column to create sequences.
MAX_NUM_WORDS = 30000
MAX_SEQ_LEN = 60  # tune depending on your dataset; tweets are short so 50-80 usually fine
EMBEDDING_DIM = 100

tokenizer = Tokenizer(num_words=MAX_NUM_WORDS, oov_token='<OOV>')
tokenizer.fit_on_texts(df['clean_text'].tolist())
sequences = tokenizer.texts_to_sequences(df['clean_text'].tolist())
word_index = tokenizer.word_index
print("Vocabulary size (word_index):", len(word_index))

X_seq = pad_sequences(sequences, maxlen=MAX_SEQ_LEN, padding='post', truncating='post')
y_labels = df['label'].values
y_cat = to_categorical(y_labels, num_classes=num_classes)

# Train/Test split for DL
X_train_seq, X_test_seq, y_train_seq, y_test_seq = train_test_split(
    X_seq, y_cat, test_size=0.2, random_state=42, stratify=y_labels
)

# For metrics we will want the string labels too
y_test_seq_raw = df['sentiment'].iloc[y_test_seq.argmax(axis=1)].values if False else None  # Not used; kept for reference

# ========================
# 11. Deep Learning Models
# ========================
# Common callbacks
es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=1)

def build_rnn_model(vocab_size, embed_dim, seq_len, num_classes):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=seq_len))
    model.add(SimpleRNN(64, return_sequences=False))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

def build_lstm_model(vocab_size, embed_dim, seq_len, num_classes, bidirectional=False):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=seq_len))
    if bidirectional:
        model.add(Bidirectional(LSTM(64)))
    else:
        model.add(LSTM(64))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

vocab_size = min(MAX_NUM_WORDS, len(word_index)+1)

# --- Simple RNN
rnn_model = build_rnn_model(vocab_size=vocab_size, embed_dim=EMBEDDING_DIM, seq_len=MAX_SEQ_LEN, num_classes=num_classes)
rnn_model.summary()
history_rnn = rnn_model.fit(
    X_train_seq, y_train_seq,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[es],
    verbose=1
)

# Evaluate RNN
rnn_pred_prob = rnn_model.predict(X_test_seq)
rnn_pred = np.argmax(rnn_pred_prob, axis=1)
rnn_pred_labels = le.inverse_transform(rnn_pred)
y_test_labels = le.inverse_transform(np.argmax(y_test_seq, axis=1))
evaluate_and_store("SimpleRNN (Keras)", y_test_labels, rnn_pred_labels, rnn_pred_prob)

# --- LSTM
lstm_model = build_lstm_model(vocab_size=vocab_size, embed_dim=EMBEDDING_DIM, seq_len=MAX_SEQ_LEN, num_classes=num_classes, bidirectional=True)
lstm_model.summary()
history_lstm = lstm_model.fit(
    X_train_seq, y_train_seq,
    validation_split=0.1,
    epochs=10,
    batch_size=64,
    callbacks=[es],
    verbose=1
)

# Evaluate LSTM
lstm_pred_prob = lstm_model.predict(X_test_seq)
lstm_pred = np.argmax(lstm_pred_prob, axis=1)
lstm_pred_labels = le.inverse_transform(lstm_pred)
y_test_labels = le.inverse_transform(np.argmax(y_test_seq, axis=1))
evaluate_and_store("BiLSTM (Keras)", y_test_labels, lstm_pred_labels, lstm_pred_prob)

# ========================
# 12. Final results summary
# ========================
final_results = pd.DataFrame(results)
# Reorder columns for readability
final_results = final_results[['Model','Accuracy','Precision','Recall','F1','AUC']]
display(final_results.sort_values('F1', ascending=False).reset_index(drop=True))

# Optionally save results
final_results.to_csv('model_results_summary.csv', index=False)
print("Results saved to model_results_summary.csv")

# ========================
# 13. Quick tips & next steps
# ========================
print("""
Next steps you may want to try:
- Hyperparameter tune your TF-IDF and ML models (GridSearchCV).
- Use pretrained embeddings (GloVe) to improve embedding quality.
- Try fine-tuning transformer models (e.g., BERT) for significant accuracy boost.
- Use stratified k-fold CV for more robust evaluation.
- Save best model(s) using model.save() (Keras) or joblib.dump (sklearn).
""")
```
